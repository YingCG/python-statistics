```python
# List of common encodings to try
file_path = "./data/billboard.csv"
encodings_to_try = ['utf-8', 'latin1', 'ISO-8859-1']

def try_encoding(file_path, encoding):
    try:
        with open(file_path, 'r', encoding=encoding) as f:
            content = f.read()
        print(f"Success! Using encoding: {encoding}")
        return True
    except UnicodeDecodeError:
        print(f"Failed with encoding: {encoding}")
        return False
    
# Try each encoding
for encoding in encodings_to_try:
    if try_encoding(file_path, encoding):
        # If successful, break out of the loop
        break
else:
    # If none of the encodings worked
    print("Unable to determine the correct encoding. Please check the file.")
```
