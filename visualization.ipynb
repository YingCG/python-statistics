```python
def plotyear(year):
    data = gapminder[gapminder.year == year]
    # bigger dot base on population of country
    area = 5e-6 * data.population
    colors = data.region.map({'Africa': 'skyblue', 'Europe':'gold', 'America': 'palegreen', 'Asia': 'coral'})
    
    data.plot.scatter('babies_per_woman', 'age5_surviving', s=area, c=colors, edgecolors='k', figsize=(12,9))
    pp.axis(ymin=50, ymax=105, xmax=8)
    pp.xlabel('babies per women')
    pp.ylabel('% children alive at 5')
plotyear(1965)
```

```python
interact(plotyear, year=widgets)
```
