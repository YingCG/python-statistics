```python
import pandas as pd
import numpy as np
%matplotlib inline

weather = pd.read_csv('./data/Auckland_CBD_-36_850774_174_764467_65af3173b13dcf00081b0b8e.csv')
weather.head()
```

```python
weather.info()
```

1. Assign a frequen of 'MS' to the datatimeindex

```python
weather['dt_iso'] = pd.to_datetime(weather['dt_iso'], format='%Y-%m-%d %H:%M:%S +0000 UTC')
weather.head()
```

```python
weather.info()
```

```python
weather.set_index('dt_iso', inplace=True)
weather.index
```

```python
weather.index.freq
```

```python
weather = weather.resample('H').first()
print("Frequency of the datetime index:", weather.index.freq)
```

```python
import matplotlib.pyplot as plt

# Plot the time series data
weather['temp'].plot(figsize=(12, 6))
plt.title('Temperature Time Series')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
weather['SMA12'] = weather['temp'].rolling(window=12).mean()
print(weather.head())
```

```python
weather['SMA12'].plot(figsize=(15, 6))
```

```python
plt.figure(figsize=(15, 6))
weather['temp'].plot(label='Temperature', color='blue')
weather['SMA12'].plot(label='12-hour SMA', color='red')
weather['rain_1h'].plot(label='Rain (1h)', color='green', linestyle='--')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Temperature, 12-hour SMA, and Rain')
plt.legend()
plt.grid(True)
plt.show()
```
