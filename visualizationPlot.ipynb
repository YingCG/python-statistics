Scatter plot of produced vs defective.

```python
import pandas as pd
%matplotlib inline

df3 = pd.read_csv('./data/df3.csv')
print(len(df3))
print(df3.head())
```

Scatter plot of produced vs defective

```python
df3.plot.scatter(x='produced', y='defective')
```

```python
df3.plot.scatter(x='produced', y='defective', c='red', figsize=(12,3), s=50)
```

histogram of produced column

```python
df3['produced'].plot.hist()
```

```python
df3['produced'].plot.hist(edgecolor='k').autoscale()
```

```python
df3['produced'].plot.hist(edgecolor='k').autoscale(axis='x', tight=True)
```

Boxplot that shows produced for each weekday

```python
df3.boxplot()
```

```python
df3.boxplot(by='weekday', figsize=(12,3))
```

```python
df3[['weekday', 'produced']].boxplot(by='weekday', figsize=(12,3))
```

KDE plot of the defective column
- In statistics, kernel density estimation is the application of kernel smoothing for probability density estimation,

```python
df3['defective'].plot.kde()
```

To increase linewidth and make linestyle dashed

```python
df3['defective'].plot.kde(ls='--', lw=5)
```

blended area plot of all the columns for just row up to 30

```python
df3.loc[0: 30].plot.area(stacked=False, alpha=0.4)
```

To move legend outside(change location)

```python
ax = df3.loc[0: 30].plot.area(stacked=False, alpha=0.4)
ax.legend(loc=0, bbox_to_anchor=(1.3, 0.5))
```
